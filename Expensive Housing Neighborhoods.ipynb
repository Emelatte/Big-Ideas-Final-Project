With the data from the 2013 American Community Survey, our group believes that the assumption of the "Best Housing" should consider the worth of the average house in each neighborhood. In most cases, the best living quarters are typically the ones more expensive due to those ones typically being the ones with the most upgrades, renovations, etc. Before we were able to narrow it down, we coded a visualization of the CSV file and then located the top 5 income brackets in order to fully narrow it down. Then, we would calculate the most expensive neighborhood by calculating how many houses run into those brackets.

```python
import pandas as pd

# Attempt to load the file from the specified directory
file_path = "/Users/emily/Documents/value.csv"
data = pd.read_csv(file_path)

# Display the first few rows to confirm successful loading
data.head(25)
```

```python
# Define columns that correspond to high-income brackets
high_income_brackets = [
    'Estimate; Total: - $300,000 to $399,999',
    'Estimate; Total: - $400,000 to $499,999',
    'Estimate; Total: - $500,000 to $749,999',
    'Estimate; Total: - $750,000 to $999,999',
    'Estimate; Total: - $1,000,000 or more'
]

# Calculate the total Most Expensive Housing estimate for each neighborhood
data['Most Expensive Housing'] = data[high_income_brackets].sum(axis=1)

# Sort neighborhoods by the Most Expensive Housing total in descending order and select the top 10
top_10_expensive_neighborhoods = data[['Neighborhood', 'Most Expensive Housing']].sort_values(by='Most Expensive Housing', ascending=False).head(10)

# Display the result
top_10_expensive_neighborhoods
```

# Our Results
According to our dataset script, the most expensive housing is in Squirrel Hill North. This could lead us to assume that Squirrel Hill is considered one of the best neighborhoods with the best houses.
