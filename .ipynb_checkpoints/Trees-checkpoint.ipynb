```python
#Imports pandas and csv files
import pandas as pd
import geopandas as gp
trees = pd.read_csv("Trees.csv", low_memory=False)
trees[["common_name", "neighborhood"]]
```

```python
#Counts how many trees are in each neighborhood
trees_per_neighborhood = {}

for index, row in trees.iterrows():
    neighborhood = row['neighborhood']

    if neighborhood in trees_per_neighborhood:
        trees_per_neighborhood[neighborhood] += 1
    else:
        trees_per_neighborhood[neighborhood] = 1
```

```python
#Calculates city average
total = 0
neighborhood_count = 0

for key in trees_per_neighborhood:
    total += trees_per_neighborhood[key]
    neighborhood_count +=1

avg = total / neighborhood_count
avg = int(avg)
print(avg)
```

```python
#Neighborhoods with tree count above average
aboveAvg = [key for key, value in trees_per_neighborhood.items() if value < avg]
for key in aboveAvg:
    del trees_per_neighborhood[key]

for key in trees_per_neighborhood:
    print(key + ":")
    print(trees_per_neighborhood[key])
```
